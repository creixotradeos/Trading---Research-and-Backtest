```python
def intraday_interval_converter():
    '''
    This function is to convert time_window into desired time_window and daily time
    As long as both factors are in the same units (minutes & minutes, hour & hour)
    '''
    FMT = '%H:%M:%S'
    dt_sliced = None; day_session_sliced = None
    
    converter = {}
    tdelta_def = datetime.datetime.strptime('16:00:00', FMT) - datetime.datetime.strptime('09:30:00', FMT)
    new_interval = int(designed_interval / interval)
    if interval == 60:
        day_session = int((tdelta_def.seconds / (interval*60)) + 1)
    else:
        day_session = int(tdelta_def.seconds / (interval*60))
    
    if sliced:
        FMT = '%H:%M:%S'
        tdelta = datetime.datetime.strptime(time_end, FMT) - datetime.datetime.strptime(time_start, FMT)
        day_session_sliced = int(tdelta.seconds / (interval*60)) # int(60 / interval * tdelta.seconds/3600)
        if type(dt['time'][0]) != datetime.time:
            dt['time'] = pd.to_datetime(dt['time']).dt.time
        dt_sliced = dt[(dt['time'] >= datetime.datetime.strptime(time_start, '%H:%M:%S').time())\
                        & (dt['time'] <= datetime.datetime.strptime(time_end, '%H:%M:%S').time())].copy()
        
    converter = {'interval': new_interval, 'daily_full': day_session, 'daily_sliced': day_session_sliced}
    return converter, dt_sliced
```

# Create Strategy

```python
class OTiCall(bt.Strategy):
    
    def log(self, txt, dt = None, h = None, doprint = True):
        '''Logging function for this strategy'''
        dt = dt or self.datas[0].datetime.date(0)
        h = h or self.datas[0].datetime.time(0)
        #h = self.datas[0].datetime.hour(0)
        print('%s %s, %s' % (dt.isoformat(), h.isoformat(), txt))      
        
    def __init__(self): 
        # Keep a reference to close price for data[0] in the series
        self.dataclose = self.datas[0].close
        self.dataopen = self.datas[0].open
        self.datavol = self.datas[0].volume
        self.datahigh = self.datas[0].high
        self.datalow = self.datas[0].low
        
        # Keep track of pending order and buy price-commission
        self.order = None
        self.price = None
        self.buycomm = None
        
        # Indicators
        ## Upper
        self.ema1 = bt.indicators.EMA(self.dataclose, period = 6)
        self.ema2 = bt.indicators.EMA(self.dataclose, period = 10)
        self.ema3 = bt.indicators.EMA(self.dataclose, period = 20)
        self.bband1 = bt.indicators.BollingerBands(self.dataclose, period = 20, devfactor = 2)
        self.delayed_bband_top = self.bband1.lines.top(-1)
        self.delayed_bband_bot = self.bband1.lines.bot(-1)
        self.bband2 = bt.indicators.BollingerBands(self.dataclose, period = 20, devfactor = 3)
        
        ## Lower 1
        self.macd = bt.indicators.MACDHisto(period_me1 = 6, period_me2 = 17, period_signal = 9)
        self.delayed_macd = self.macd.lines.histo(-1)
        
        ## Lower 2
        self.adx = bt.indicators.AverageDirectionalMovementIndex(period = 5)
        sig = self.adx(-1)
        self.sig = self.adx.lines.adx - sig
        
        ## Lower 3
        self.stoch = bt.indicators.Stochastic(period = 8, period_dfast = 2, period_dslow = 2)
        self.delayed_stochK = self.stoch.lines.percK(-1); self.delayed_stochK2 = self.stoch.lines.percK(-2)
        self.delayed_stochD = self.stoch.lines.percD(-1); self.delayed_stochD2 = self.stoch.lines.percD(-2)
        self.cross_stoch = bt.indicators.CrossUp(self.stoch.lines.percK, self.stoch.lines.percD)
        self.cross_stoch2 = bt.indicators.CrossUp(self.delayed_stochK, self.delayed_stochD)
        self.cross_stoch3 = bt.indicators.CrossUp(self.delayed_stochK2, self.delayed_stochD2)
        
    def notify_order(self, order):
        # If Buy and/or Sell submitted then accepted by broker - Nothing to do
        if order.status in [order.Submitted, order.Accepted]:
            return
        
        # Check if an order is completed
        if order.status in [order.Completed]:
            
            if order.isbuy():
                self.log('BUY EXECUTED, Price: {:,.2f}'.format(order.executed.price))
                self.target = order.executed.price * 1.001
            
            elif order.issell():
                self.log('SELL EXECUTED, Price: {:,.2f}'.format(order.executed.price))              
        elif order.status in [order.Cancelled, order.Margin, order.Rejected]:
            self.log('ORDER CANCELLED')
            
        # Write down that there is no order
        self.order = None
        return self.target
        
    def notify_trade(self, trade):
        if not trade.isclosed:
            return
        
        self.log('OPERATION PROFIT/LOSS, GROSS: {:,.2f}'.format(trade.pnl))
    
    def next(self):
        # Log the next price
        self.log('Close: {:,.2f}'.format(self.dataclose[0]))
        
        # If there is a pending order do not order another one
        if self.order:
            return
        
        # Check if we are in the market
        if not self.position:
            # Not in the market
            
            # Buy if
            def buyif():            
                rules1 = [(self.dataclose[0] >= self.ema1[0] or self.dataopen[0] >= self.ema1[0]),
                          (self.dataclose[0] >= self.ema2[0] or self.dataopen[0] >= self.ema2[0]),
                          (self.dataclose[0] >= self.ema3[0] or self.dataopen[0] >= self.ema3[0]),
                         ]
                rules2 = [(self.macd.lines.histo[0] > self.delayed_macd[0]), # Higher current MACD
                          (self.macd.lines.histo[0] > 0.2), # MACD above 0.2
                          (abs(self.macd.lines.histo[0]) > 0.1), # MACD's abs value above 0.1
                         ]
                rules3 = [(self.adx.lines.adx[0] >= 20 and self.adx.lines.adx[0] <= 50 ),] # ADX is between range]
                rules4 = [(self.cross_stoch.lines.cross == True or self.cross_stoch2.lines.cross == True \
                           or self.cross_stoch3.lines.cross == True)] # Crossing of StochK and StochD
                rules = [rules1, rules2, rules3, rules4]

                return rules

            if all(buyif()):
                self.log('BUY CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                self.order = self.buy()
        
        else:
        # Already in the market
            
        # Sell if
#             rules1 = [(self.dataclose[0] >= self.bband1.lines.top[0])]
#             rules2 = [(self.dataopen[0] >= self.bband1.lines.top[0])]
#             rules3 = [(self.datahigh[0] >= self.bband1.lines.top[0])]
#             rules4 = [(self.datalow[0] >= self.bband1.lines.top[0])]
            sellif = [self.dataclose[0] >= self.target]
            if all(sellif):# or all(rules2):
                self.log('SELL CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                self.order = self.sell()
```

```python
class OTi_Full(bt.Strategy):
    params = (('ema1', 6), ('ema2', 10), ('ema3', 20),
              ('bband1_period', 20), ('bband1_stdev', 2), ('bband2_period', 20), ('bband2_stdev', 3),
              ('macd_me1', 6), ('macd_me2', 17), ('macd_sign', 9),
              ('adx_period', 5), ('stoch_period', 8), ('dfast', 2), ('dslow', 2), 
              ('target', 0.01))
    identity = params
    def log(self, txt, dt = None, h = None, doprint = True):
        '''Logging function for this strategy'''
        dt = dt or self.datas[0].datetime.date(0)
        h = h or self.datas[0].datetime.time(0)
        print('%s %s, %s' % (dt.isoformat(), h.isoformat(), txt))
    def __init__(self):
        if True:# Keep a reference to prices for data[0] in the series
            self.dataclose = self.datas[0].close
            self.dataopen = self.datas[0].open
            self.datavol = self.datas[0].volume
            self.datahigh = self.datas[0].high
            self.datalow = self.datas[0].low        
        if True:# Keep track of pending order, buy price-commission, and call-put
            self.order = None
            self.price = None
            self.buycomm = None
            self.call_put = None 
        # Indicators
        if True:## Upper - EMA and Bollinger
            self.ema1 = bt.indicators.EMA(self.dataclose, period = self.p.ema1)
            self.ema2 = bt.indicators.EMA(self.dataclose, period = self.p.ema2)
            self.ema3 = bt.indicators.EMA(self.dataclose, period = self.p.ema3)
            self.bband1 = bt.indicators.BollingerBands(self.dataclose, period = self.p.bband1_period, devfactor = self.p.bband1_stdev)
            self.delayed_bband_top = self.bband1.lines.top(-1)
            self.delayed_bband_bot = self.bband1.lines.bot(-1)
            self.bband2 = bt.indicators.BollingerBands(self.dataclose, period = self.p.bband2_period, devfactor = self.p.bband2_stdev)
        if True:## Lower 1 - MACD
            self.macd = bt.indicators.MACDHisto(period_me1 = self.p.macd_me1, period_me2 = self.p.macd_me2, period_signal = self.p.macd_sign)
            self.delayed_macd = self.macd.lines.histo(-1)
        if True:## Lower 2 - ADX
            self.adx = bt.indicators.AverageDirectionalMovementIndex(period = self.p.adx_period)
            sig = self.adx(-1)
            self.sig = self.adx.lines.adx - sig
        if True:## Lower 3 - Stochastic
            self.stoch = bt.indicators.Stochastic(period = self.p.stoch_period, period_dfast = self.p.dfast, period_dslow = self.p.dslow)
            self.delayed_stochK = self.stoch.lines.percK(-1); self.delayed_stochK2 = self.stoch.lines.percK(-2)
            self.delayed_stochD = self.stoch.lines.percD(-1); self.delayed_stochD2 = self.stoch.lines.percD(-2)
            self.cross_stoch = bt.indicators.CrossOver(self.stoch.lines.percK, self.stoch.lines.percD)
            self.cross_stoch2 = bt.indicators.CrossOver(self.delayed_stochK, self.delayed_stochD)
            self.cross_stoch3 = bt.indicators.CrossOver(self.delayed_stochK2, self.delayed_stochD2)
    def next(self):
        # Log the next price
        self.log('Close: {:,.2f}'.format(self.dataclose[0]))
        # If there is a pending order, do not order another one
        if self.order:
            return      
        # This is for one position at a time
        # Check if there is a position in the market
        if not self.position: # Not in the market, open position
            def call_if():
                # EMA
                call_rule1 = [(self.dataclose[0] > self.ema1[0] or self.dataopen[0] > self.ema1[0]),
                              (self.dataclose[0] > self.ema2[0] or self.dataopen[0] > self.ema2[0]),
                              (self.dataclose[0] > self.ema3[0] or self.dataopen[0] > self.ema3[0]),
                             ]
                # MACD
                call_rule2 = [(self.macd.lines.histo[0] > self.delayed_macd[0]),
                              (self.macd.lines.histo[0] > -0.2 and self.macd.lines.histo[0] < 0.2),
                              (abs(self.macd.lines.histo[0]) > 0.1),
                             ]
                # ADX
                call_rule3 = [(self.adx.lines.adx[0] >= 20 and self.adx.lines.adx[0] <= 50),
                             ]
                # Stochastic
                call_rule4 = [(self.cross_stoch.lines.crossover == 1 or self.cross_stoch2.lines.crossover == 1\
                              or self.cross_stoch3.lines.crossover == 1),
                             ]
                # Summary
                sum_rules = [call_rule1, call_rule2, call_rule3, call_rule4]
                call_rules = []
                for rules in sum_rules:
                    correct = all(rules)
                    call_rules.append(correct)
                return call_rules
            def put_if():
                # EMA
                put_rule1 = [(self.dataclose[0] < self.ema1[0] or self.dataopen[0] < self.ema1),
                             (self.dataclose[0] < self.ema2[0] or self.dataopen[0] < self.ema2),
                             (self.dataclose[0] < self.ema3[0] or self.dataopen[0] < self.ema3),
                            ]
                # MACD
                put_rule2 = [(self.macd.lines.histo[0] < self.delayed_macd[0]),
                              (self.macd.lines.histo[0] > -0.2),
                              (abs(self.macd.lines.histo[0]) > 0.1),
                             ]
                # ADX
                put_rule3 = [(self.adx.lines.adx[0] >= 20 and self.adx.lines.adx[0] <= 50),
                            ]
                # Stochastic
                put_rule4 = [(self.cross_stoch.lines.crossover == -1 or self.cross_stoch2.lines.crossover == -1\
                              or self.cross_stoch3.lines.crossover == -1),
                             (self.stoch.lines.percD[0] > self.stoch.lines.percK[0]),
                             (self.stoch.lines.percD[0] >= 50 or self.delayed_stochD >= 50 or self.delayed_stochD2 >= 50),
                             (self.stoch.lines.percK[0] >= 50 or self.delayed_stochK >= 50 or self.delayed_stochK2 >= 50),
                             ]                
                # Summary
                sum_rules = [put_rule1, put_rule2, put_rule3, put_rule4]
                put_rules = []
                for rules in sum_rules:
                    correct = all(rules)
                    put_rules.append(correct)
                return put_rules
            if all(call_if()): 
                self.log('BUY CALL CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                self.order = self.buy()
                self.call_put = 'buy call'
            elif all(put_if()):
                self.log('BUY PUT CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                self.order = self.sell()
                self.call_put = 'buy put'            
        else: # Close position
            #if self.datas[0].datetime.time(0) < datetime.time(15, 30): # time
            target_achieved_call = [((self.target_p <= self.dataclose[0]) or (self.target_p <= self.dataopen[0])) or
                                    ((self.target_l >= self.dataclose[0]) or (self.target_l >= self.dataopen[0]))]
            target_achieved_put = [((self.target_p >= self.dataclose[0]) or (self.target_p >= self.dataopen[0])) or
                                   ((self.target_l <= self.dataclose[0]) or (self.target_l <= self.dataopen[0]))]
            if  all(target_achieved_call) and self.call_put == 'buy call':
                self.log('SELL CALL CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                self.order = self.sell()
                self.call_put = 'sell call'
            elif all(target_achieved_put) and self.call_put == 'buy put':
                self.log('SELL PUT CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                self.order = self.buy()
                self.call_put = 'sell put'        
    def notify_order(self, order):
        # If Buy and/or Sell submitted then accepted by broker - Nothing to do
        if order.status in [order.Submitted, order.Accepted]:
            return
        #Check if an order is completed
        if order.status in [order.Completed]:
            if self.call_put == 'buy call' and order.isbuy():
                self.log('BUY CALL EXECUTED, Price: {:,.2f}'.format(order.executed.price))
                self.target_p = order.executed.price * (1+self.p.target)
                self.target_l = order.executed.price * (1-self.p.target*10)
            elif self.call_put == 'sell call' and order.issell():
                self.log('SELL CALL EXECUTED, Price: {:,.2f}'.format(order.executed.price))
            elif self.call_put == 'buy put' and order.issell():
                self.log('BUY PUT EXECUTED, Price: {:,.2f}'.format(order.executed.price))
                self.target_p = order.executed.price * (1-self.p.target)
                self.target_l = order.executed.price * (1+self.p.target)
            elif self.call_put == 'sell put' and order.isbuy():
                self.log('SELL PUT EXECUTED, Price: {:,.2f}'.format(order.executed.price))
        elif order.status in [order.Cancelled, order.Margin, order.Rejected]:
            self.log('ORDER CANCELLED')
        # Write down that there is no order
        self.order = None
        return self.target_p, self.target_l
    def notify_trade(self, trade):
        if not trade.isclosed:
            return        
        self.log('OPERATION PROFIT/LOSS, GROSS: {:,.2f}'.format(trade.pnl))
```

```python
class WilliamR(bt.Strategy):
    params = (('target', 0.005), ('printlog', False), ('use_time_limit', True),
              ('william_period', 9), ('william_lower', -80), ('william_upper', -20),
              ('bband1_period', 20), ('bband1_stdev', 2), ('bband2_period', 20), ('bband2_stdev', 3),
              ('macd_me1', 6), ('macd_me2', 17), ('macd_sign', 9),
              ('adx_period', 5),
             )
    identity = params
    def log(self, txt, dt = None, h = None, doprint = False):
        '''Logging function for this strategy'''
        dt = dt or self.datas[0].datetime.date(0)
        h = h or self.datas[0].datetime.time(0)
        if self.params.printlog or doprint:
            print('%s %s, %s' % (dt.isoformat(), h.isoformat(), txt))
    def __init__(self):
        if 1:# Keep a reference to prices for data[0] in the series
            self.dataclose = self.datas[0].close
            self.dataopen = self.datas[0].open
            self.datavol = self.datas[0].volume
            self.datahigh = self.datas[0].high
            self.datalow = self.datas[0].low        
        if 1:# Keep track of pending order, buy price-commission, and call-put
            self.order = None
            self.price = None
            self.buycomm = None
            self.call_put = None 
        #----------------------Indicators----------------------#
        if 1: # William %R
            self.williamR = bt.indicators.WilliamsR(period = self.p.william_period, upperband = self.p.william_upper, 
                                                    lowerband = self.p.william_lower)
        if 1: # Support-Resistance
            self.support_resist = bt.indicators.PivotPoint(self.data1, _autoplot = True)
        if 1: # Bollinger Band
            self.bband1 = bt.indicators.BollingerBands(self.dataclose, 
                                                       period = self.p.bband1_period, devfactor = self.p.bband1_stdev)
            self.delayed_bband_top = self.bband1.lines.top(-1)
            self.delayed_bband_bot = self.bband1.lines.bot(-1)
            self.bband2 = bt.indicators.BollingerBands(self.dataclose, 
                                                       period = self.p.bband2_period, devfactor = self.p.bband2_stdev)            
        if 1: # MACD
            self.macd = bt.indicators.MACDHisto(period_me1 = self.p.macd_me1, period_me2 = self.p.macd_me2, period_signal = self.p.macd_sign)
            self.delayed_macd = self.macd.lines.histo(-1)            
        if 1: # ADX
            self.adx = bt.indicators.AverageDirectionalMovementIndex(period = self.p.adx_period)
            self.delayed_adx = self.adx.lines.adx(-1)
            sig = self.adx(-1)
            self.sig = self.adx.lines.adx - sig
    def next(self):
        # Log the next price
        self.log('Close: {:,.2f}'.format(self.dataclose[0]))
        # If there is a pending order, do not order another one
        if self.order:
            return      
        # This is for one position at a time
        # Check if there is a position in the market
        if not self.position: # Not in the market, open position
            time_limit = self.datas[0].datetime.time(0) < datetime.time(13, 0)
            def call_if():
                #----------------------William%R----------------------#
                call_rule1 = [self.williamR.lines.percR[0] > self.p.william_upper]
                #----------------------Support-Resistance----------------------#
                call_rule2 = [self.dataclose[0] > self.support_resist.lines.r1]
                #----------------------Bollinger Band----------------------#
                call_rule3 = [(self.bband1.lines.top[0] > self.delayed_bband_top[0]),
                              (self.bband1.lines.bot[0] < self.delayed_bband_bot[0]),
                             ]
                #----------------------MACD----------------------#               
                call_rule4 = [(self.macd.lines.histo[0] > self.delayed_macd[0]),
                              #(self.macd.lines.histo[0] > -0.2 and self.macd.lines.histo[0] < 0.2),
                              #(abs(self.macd.lines.histo[0]) > 0.1),
                             ]
                #----------------------ADX----------------------#
                call_rule5 = [(self.adx.lines.adx[0] > 20 and self.adx.lines.adx[0] <= 50),
                             (self.adx.lines.adx[0] > self.delayed_adx[0]),
                            ]
                # Summary
                sum_rules = [call_rule1, call_rule2, call_rule3, call_rule4, call_rule5]
                #sum_rules = [call_rule5]
                call_rules = []
                for rules in sum_rules:
                    correct = all(rules)
                    call_rules.append(correct)
                return call_rules
            def put_if():
                #----------------------William%R----------------------#
                put_rule1 = [self.williamR.lines.percR[0] < self.p.william_lower]
                #----------------------Suppport-Resistance----------------------#
                put_rule2 = [self.dataclose[0] < self.support_resist.lines.s1]
                #----------------------Bollinger Band----------------------#
                put_rule3 = [(self.bband1.lines.top[0] > self.delayed_bband_top[0]),
                             (self.bband1.lines.bot[0] < self.delayed_bband_bot[0]),
                            ]
                #----------------------MACD----------------------#
                put_rule4 = [(self.macd.lines.histo[0] < self.delayed_macd[0]),
                             #(self.macd.lines.histo[0] > -0.2),
                             #(abs(self.macd.lines.histo[0]) > 0.1),
                            ]
                #----------------------ADX----------------------#
                put_rule5 = [(self.adx.lines.adx[0] > 20 and self.adx.lines.adx[0] <= 25),
                             (self.adx.lines.adx[0] > self.delayed_adx[0]),
                            ]
                # Summary
                sum_rules = [put_rule1, put_rule2, put_rule3, put_rule4, put_rule5]
                #sum_rules = [put_rule5]
                put_rules = []
                for rules in sum_rules:
                    correct = all(rules)
                    put_rules.append(correct)
                return put_rules
            if self.p.use_time_limit and time_limit:           
                if all(call_if()): 
                    self.log('BUY CALL CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                    self.order = self.buy()
                    self.call_put = 'buy call'
                elif all(put_if()):
                    self.log('BUY PUT CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                    self.order = self.sell()
                    self.call_put = 'buy put'            
            elif not self.p.use_time_limit:
                if all(call_if()): 
                    self.log('BUY CALL CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                    self.order = self.buy()
                    self.call_put = 'buy call'
                elif all(put_if()):
                    self.log('BUY PUT CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                    self.order = self.sell()
                    self.call_put = 'buy put'            
        else: # Close position
            time_limit = self.datas[0].datetime.time(0) < datetime.time(15, 0)
            if self.p.use_time_limit and time_limit: # time
                if 1: # default
                    target_achieved_call = [((self.target_p <= self.dataclose[0]) or (self.target_p <= self.dataopen[0])) or
                                            ((self.target_l >= self.dataclose[0]) or (self.target_l >= self.dataopen[0]))]
                    target_achieved_put = [((self.target_p >= self.dataclose[0]) or (self.target_p >= self.dataopen[0])) or
                                           ((self.target_l <= self.dataclose[0]) or (self.target_l <= self.dataopen[0]))]
                    if  all(target_achieved_call) and self.call_put == 'buy call':
                        self.log('SELL CALL CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                        self.order = self.sell()
                        self.call_put = 'sell call'
                    elif all(target_achieved_put) and self.call_put == 'buy put':
                        self.log('SELL PUT CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                        self.order = self.buy()
                        self.call_put = 'sell put'        
                if 0: # reverse rule
                    def call_if():
                        #----------------------William%R----------------------#
                        call_rule1 = [self.williamR.lines.percR[0] > self.p.william_upper]
                        #----------------------Support-Resistance----------------------#
                        call_rule2 = [self.dataclose[0] > self.support_resist.lines.r1]
                        #----------------------Bollinger Band----------------------#
                        call_rule3 = [(self.bband1.lines.top[0] > self.delayed_bband_top[0]),
                                      (self.bband1.lines.bot[0] < self.delayed_bband_bot[0]),
                                     ]
                        #----------------------MACD----------------------#               
                        call_rule4 = [(self.macd.lines.histo[0] > self.delayed_macd[0]),
                                      #(self.macd.lines.histo[0] > -0.2 and self.macd.lines.histo[0] < 0.2),
                                      #(abs(self.macd.lines.histo[0]) > 0.1),
                                     ]
                        #----------------------ADX----------------------#
                        call_rule5 = [(self.adx.lines.adx[0] > 20 and self.adx.lines.adx[0] <= 50),
                                     (self.adx.lines.adx[0] > self.delayed_adx[0]),
                                    ]
                        # Summary
                        sum_rules = [call_rule1, call_rule2, call_rule3, call_rule4, call_rule5]
                        #sum_rules = [call_rule5]
                        call_rules = []
                        for rules in sum_rules:
                            correct = all(rules)
                            call_rules.append(correct)
                        return call_rules
                    def put_if():
                        #----------------------William%R----------------------#
                        put_rule1 = [self.williamR.lines.percR[0] < self.p.william_lower]
                        #----------------------Suppport-Resistance----------------------#
                        put_rule2 = [self.dataclose[0] < self.support_resist.lines.s1]
                        #----------------------Bollinger Band----------------------#
                        put_rule3 = [(self.bband1.lines.top[0] > self.delayed_bband_top[0]),
                                     (self.bband1.lines.bot[0] < self.delayed_bband_bot[0]),
                                    ]
                        #----------------------MACD----------------------#
                        put_rule4 = [(self.macd.lines.histo[0] < self.delayed_macd[0]),
                                     #(self.macd.lines.histo[0] > -0.2),
                                     #(abs(self.macd.lines.histo[0]) > 0.1),
                                    ]
                        #----------------------ADX----------------------#
                        put_rule5 = [(self.adx.lines.adx[0] > 20 and self.adx.lines.adx[0] <= 25),
                                     (self.adx.lines.adx[0] > self.delayed_adx[0]),
                                    ]
                        # Summary
                        sum_rules = [put_rule1, put_rule2, put_rule3, put_rule4, put_rule5]
                        #sum_rules = [put_rule5]
                        put_rules = []
                        for rules in sum_rules:
                            correct = all(rules)
                            put_rules.append(correct)
                        return put_rules
                    if all(put_if()) and self.call_put == 'buy call': 
                        self.log('SELL CALL CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                        self.order = self.sell()
                        self.call_put = 'sell call'
                    elif all(call_if()) and self.call_put == 'buy put':
                        self.log('SELL PUT CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                        self.order = self.buy()
                        self.call_put = 'sell put'      
            elif not self.p.use_time_limit:
                if 1: # default
                    target_achieved_call = [((self.target_p <= self.dataclose[0]) or (self.target_p <= self.dataopen[0])) or
                                            ((self.target_l >= self.dataclose[0]) or (self.target_l >= self.dataopen[0]))]
                    target_achieved_put = [((self.target_p >= self.dataclose[0]) or (self.target_p >= self.dataopen[0])) or
                                           ((self.target_l <= self.dataclose[0]) or (self.target_l <= self.dataopen[0]))]
                    if  all(target_achieved_call) and self.call_put == 'buy call':
                        self.log('SELL CALL CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                        self.order = self.sell()
                        self.call_put = 'sell call'
                    elif all(target_achieved_put) and self.call_put == 'buy put':
                        self.log('SELL PUT CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                        self.order = self.buy()
                        self.call_put = 'sell put'        
                if 0: # reverse rule
                    def call_if():
                        #----------------------William%R----------------------#
                        call_rule1 = [self.williamR.lines.percR[0] > self.p.william_upper]
                        #----------------------Support-Resistance----------------------#
                        call_rule2 = [self.dataclose[0] > self.support_resist.lines.r1]
                        #----------------------Bollinger Band----------------------#
                        call_rule3 = [(self.bband1.lines.top[0] > self.delayed_bband_top[0]),
                                      (self.bband1.lines.bot[0] < self.delayed_bband_bot[0]),
                                     ]
                        #----------------------MACD----------------------#               
                        call_rule4 = [(self.macd.lines.histo[0] > self.delayed_macd[0]),
                                      #(self.macd.lines.histo[0] > -0.2 and self.macd.lines.histo[0] < 0.2),
                                      #(abs(self.macd.lines.histo[0]) > 0.1),
                                     ]
                        #----------------------ADX----------------------#
                        call_rule5 = [(self.adx.lines.adx[0] > 20 and self.adx.lines.adx[0] <= 50),
                                     (self.adx.lines.adx[0] > self.delayed_adx[0]),
                                    ]
                        # Summary
                        sum_rules = [call_rule1, call_rule2, call_rule3, call_rule4, call_rule5]
                        #sum_rules = [call_rule5]
                        call_rules = []
                        for rules in sum_rules:
                            correct = all(rules)
                            call_rules.append(correct)
                        return call_rules
                    def put_if():
                        #----------------------William%R----------------------#
                        put_rule1 = [self.williamR.lines.percR[0] < self.p.william_lower]
                        #----------------------Suppport-Resistance----------------------#
                        put_rule2 = [self.dataclose[0] < self.support_resist.lines.s1]
                        #----------------------Bollinger Band----------------------#
                        put_rule3 = [(self.bband1.lines.top[0] > self.delayed_bband_top[0]),
                                     (self.bband1.lines.bot[0] < self.delayed_bband_bot[0]),
                                    ]
                        #----------------------MACD----------------------#
                        put_rule4 = [(self.macd.lines.histo[0] < self.delayed_macd[0]),
                                     #(self.macd.lines.histo[0] > -0.2),
                                     #(abs(self.macd.lines.histo[0]) > 0.1),
                                    ]
                        #----------------------ADX----------------------#
                        put_rule5 = [(self.adx.lines.adx[0] > 20 and self.adx.lines.adx[0] <= 25),
                                     (self.adx.lines.adx[0] > self.delayed_adx[0]),
                                    ]
                        # Summary
                        sum_rules = [put_rule1, put_rule2, put_rule3, put_rule4, put_rule5]
                        #sum_rules = [put_rule5]
                        put_rules = []
                        for rules in sum_rules:
                            correct = all(rules)
                            put_rules.append(correct)
                        return put_rules
                    if all(put_if()) and self.call_put == 'buy call': 
                        self.log('SELL CALL CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                        self.order = self.sell()
                        self.call_put = 'sell call'
                    elif all(call_if()) and self.call_put == 'buy put':
                        self.log('SELL PUT CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                        self.order = self.buy()
                        self.call_put = 'sell put'                      
    def notify_order(self, order):
        # If Buy and/or Sell submitted then accepted by broker - Nothing to do
        if order.status in [order.Submitted, order.Accepted]:
            return
        #Check if an order is completed
        if order.status in [order.Completed]:
            if self.call_put == 'buy call' and order.isbuy():
                self.log('BUY CALL EXECUTED, Price: {:,.2f}'.format(order.executed.price))
                self.target_p = order.executed.price * (1+self.p.target)
                self.target_l = order.executed.price * (1-self.p.target*10)
            elif self.call_put == 'sell call' and order.issell():
                self.log('SELL CALL EXECUTED, Price: {:,.2f}'.format(order.executed.price))
            elif self.call_put == 'buy put' and order.issell():
                self.log('BUY PUT EXECUTED, Price: {:,.2f}'.format(order.executed.price))
                self.target_p = order.executed.price * (1-self.p.target)
                self.target_l = order.executed.price * (1+self.p.target)
            elif self.call_put == 'sell put' and order.isbuy():
                self.log('SELL PUT EXECUTED, Price: {:,.2f}'.format(order.executed.price))
        elif order.status in [order.Cancelled, order.Margin, order.Rejected]:
            self.log('ORDER CANCELLED', doprint = self.p.print)
        # Write down that there is no order
        self.order = None
        return self.target_p, self.target_l
    def notify_trade(self, trade):
        if not trade.isclosed:
            return        
        self.log('OPERATION PROFIT/LOSS, GROSS: {:,.2f}'.format(trade.pnl))
```

```python
class OTi_Full2(bt.Strategy):
    params = (('ema1', 6), ('ema2', 10), ('ema3', 20),
              ('bband1_period', 20), ('bband1_stdev', 2), ('bband2_period', 20), ('bband2_stdev', 3),
              ('macd_me1', 6), ('macd_me2', 17), ('macd_sign', 9),
              ('adx_period', 5), ('stoch_period', 8), ('dfast', 2), ('dslow', 2), 
              ('target', 0.01))
    identity = params
    def log(self, txt, dt = None, h = None, doprint = True):
        '''Logging function for this strategy'''
        dt = dt or self.datas[0].datetime.date(0)
        h = h or self.datas[0].datetime.time(0)
        print('%s %s, %s' % (dt.isoformat(), h.isoformat(), txt))
    def __init__(self):
        if True:# Keep a reference to prices for data[0] in the series
            self.dataclose = self.datas[0].close
            self.dataopen = self.datas[0].open
            self.datavol = self.datas[0].volume
            self.datahigh = self.datas[0].high
            self.datalow = self.datas[0].low        
        if True:# Keep track of pending order, buy price-commission, and call-put
            self.order = None
            self.price = None
            self.buycomm = None
            self.call_put = None 
        # Indicators
        if True:## Upper - EMA and Bollinger
            self.ema1 = bt.indicators.EMA(self.dataclose, period = self.p.ema1)
            self.ema2 = bt.indicators.EMA(self.dataclose, period = self.p.ema2)
            self.ema3 = bt.indicators.EMA(self.dataclose, period = self.p.ema3)
            self.bband1 = bt.indicators.BollingerBands(self.dataclose, period = self.p.bband1_period, devfactor = self.p.bband1_stdev)
            self.delayed_bband_top = self.bband1.lines.top(-1)
            self.delayed_bband_bot = self.bband1.lines.bot(-1)
            self.bband2 = bt.indicators.BollingerBands(self.dataclose, period = self.p.bband2_period, devfactor = self.p.bband2_stdev)
        if True:## Lower 1 - MACD
            self.macd = bt.indicators.MACDHisto(period_me1 = self.p.macd_me1, period_me2 = self.p.macd_me2, period_signal = self.p.macd_sign)
            self.delayed_macd = self.macd.lines.histo(-1)
        if True:## Lower 2 - ADX
            self.adx = bt.indicators.AverageDirectionalMovementIndex(period = self.p.adx_period)
            sig = self.adx(-1)
            self.sig = self.adx.lines.adx - sig
        if True:## Lower 3 - Stochastic
            self.stoch = bt.indicators.Stochastic(period = self.p.stoch_period, period_dfast = self.p.dfast, period_dslow = self.p.dslow)
            self.delayed_stochK = self.stoch.lines.percK(-1); self.delayed_stochK2 = self.stoch.lines.percK(-2)
            self.delayed_stochD = self.stoch.lines.percD(-1); self.delayed_stochD2 = self.stoch.lines.percD(-2)
            self.cross_stoch = bt.indicators.CrossOver(self.stoch.lines.percK, self.stoch.lines.percD)
            self.cross_stoch2 = bt.indicators.CrossOver(self.delayed_stochK, self.delayed_stochD)
            self.cross_stoch3 = bt.indicators.CrossOver(self.delayed_stochK2, self.delayed_stochD2)
    def next(self):
        # Log the next price
        self.log('Close: {:,.2f}'.format(self.dataclose[0]))
        # If there is a pending order, do not order another one
        if self.order:
            return      
        # This is for one position at a time
        # Check if there is a position in the market
        def call_if():
                # EMA
                call_rule1 = [(self.dataclose[0] > self.ema1[0] or self.dataopen[0] > self.ema1[0]),
                              (self.dataclose[0] > self.ema2[0] or self.dataopen[0] > self.ema2[0]),
                              (self.dataclose[0] > self.ema3[0] or self.dataopen[0] > self.ema3[0]),
                             ]
                # MACD
                call_rule2 = [(self.macd.lines.histo[0] > self.delayed_macd[0]),
                              (self.macd.lines.histo[0] > -0.2 and self.macd.lines.histo[0] < 0.2),
                              (abs(self.macd.lines.histo[0]) > 0.1),
                             ]
                # ADX
                call_rule3 = [(self.adx.lines.adx[0] >= 20 and self.adx.lines.adx[0] <= 50),
                             ]
                # Stochastic
                call_rule4 = [(self.cross_stoch.lines.crossover == 1 or self.cross_stoch2.lines.crossover == 1\
                              or self.cross_stoch3.lines.crossover == 1),
                             ]
                # Summary
                sum_rules = [call_rule1, call_rule2, call_rule3, call_rule4]
                call_rules = []
                for rules in sum_rules:
                    correct = all(rules)
                    call_rules.append(correct)
                return call_rules
        def put_if():
                # EMA
                put_rule1 = [(self.dataclose[0] < self.ema1[0] or self.dataopen[0] < self.ema1),
                             (self.dataclose[0] < self.ema2[0] or self.dataopen[0] < self.ema2),
                             (self.dataclose[0] < self.ema3[0] or self.dataopen[0] < self.ema3),
                            ]
                # MACD
                put_rule2 = [(self.macd.lines.histo[0] < self.delayed_macd[0]),
                              (self.macd.lines.histo[0] > -0.2),
                              (abs(self.macd.lines.histo[0]) > 0.1),
                             ]
                # ADX
                put_rule3 = [(self.adx.lines.adx[0] >= 20 and self.adx.lines.adx[0] <= 50),
                            ]
                # Stochastic
                put_rule4 = [(self.cross_stoch.lines.crossover == -1 or self.cross_stoch2.lines.crossover == -1\
                              or self.cross_stoch3.lines.crossover == -1),
                             (self.stoch.lines.percD[0] > self.stoch.lines.percK[0]),
                             (self.stoch.lines.percD[0] >= 50 or self.delayed_stochD >= 50 or self.delayed_stochD2 >= 50),
                             (self.stoch.lines.percK[0] >= 50 or self.delayed_stochK >= 50 or self.delayed_stochK2 >= 50),
                             ]                
                # Summary
                sum_rules = [put_rule1, put_rule2, put_rule3, put_rule4]
                put_rules = []
                for rules in sum_rules:
                    correct = all(rules)
                    put_rules.append(correct)
                return put_rules
        if not self.position: # Not in the market, open position
            if all(call_if()): 
                self.log('BUY CALL CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                self.order = self.buy()
                self.call_put = 'buy call'
            elif all(put_if()):
                self.log('BUY PUT CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                self.order = self.sell()
                self.call_put = 'buy put'            
        else:# Close position
            target_achieved_call = [(self.target <= self.dataclose[0]) or (self.target <= self.dataopen[0])]
            target_achieved_put = [(self.target >= self.dataclose[0]) or (self.target >= self.dataopen[0])]
            if  all(target_achieved_call) and self.call_put == 'buy call':
                self.log('SELL CALL CREATE ORDER: {:,.2f}'.format(self.target))
                self.order = self.sell()
                self.call_put = 'sell call'
            elif all(target_achieved_put) and self.call_put == 'buy put':
                self.log('SELL PUT CREATE ORDER: {:,.2f}'.format(self.target))
                self.order = self.buy()
                self.call_put = 'sell put'        
    def cancel(order):
        rule = [(self.target <= self.dataclose[0] and order.issell()) or
                (self.target >= self.dataclose[0] and order.isbuy())]
        if all(rule):
            self.cancel(order)
    def notify_order(self, order):
        # If Buy and/or Sell submitted then accepted by broker - Nothing to do
        if order.status in [order.Submitted, order.Accepted]:
            return
        #Check if an order is completed
        if order.status in [order.Completed]:
            if self.call_put == 'buy call' and order.isbuy():
                self.log('BUY CALL EXECUTED, Price: {:,.2f}'.format(order.executed.price))
                self.target = order.executed.price * (1+self.p.target)
            elif self.call_put == 'sell call' and order.issell():
                self.log('SELL CALL EXECUTED, Price: {:,.2f}'.format(order.executed.price))
            elif self.call_put == 'buy put' and order.issell():
                self.log('BUY PUT EXECUTED, Price: {:,.2f}'.format(order.executed.price))
                self.target = order.executed.price * (1-self.p.target)
            elif self.call_put == 'sell put' and order.isbuy():
                self.log('SELL PUT EXECUTED, Price: {:,.2f}'.format(order.executed.price))
        elif order.status in [order.Cancelled, order.Margin, order.Rejected]:
            self.log('ORDER CANCELLED')
        # Write down that there is no order
        self.order = None
        return self.target
    def notify_trade(self, trade):
        if not trade.isclosed:
            return        
        self.log('OPERATION PROFIT/LOSS, GROSS: {:,.2f}'.format(trade.pnl))
```

```python
class OTi_Put(bt.Strategy):
    def log(self, txt, dt = None, h = None, doprint = True):
        '''Logging function for this strategy'''
        dt = dt or self.datas[0].datetime.date(0)
        h = h or self.datas[0].datetime.time(0)
        print('%s %s, %s' % (dt.isoformat(), h.isoformat(), txt))              
    def __init__(self): 
        # Keep a reference to close price for data[0] in the series
        self.dataclose = self.datas[0].close
        self.dataopen = self.datas[0].open
        self.datavol = self.datas[0].volume
        self.datahigh = self.datas[0].high
        self.datalow = self.datas[0].low
        self.time = self.datas[0].datetime.time(0)
        # Keep track of pending order, buy price-commission, and call-put
        self.order = None
        self.price = None
        self.buycomm = None
        self.call_put = None 
        # Indicators
        ## Upper - EMA and Bollinger
        self.ema1 = bt.indicators.EMA(self.dataclose, period = 6)
        self.ema2 = bt.indicators.EMA(self.dataclose, period = 10)
        self.ema3 = bt.indicators.EMA(self.dataclose, period = 20)
        self.bband1 = bt.indicators.BollingerBands(self.dataclose, period = 20, devfactor = 2)
        self.delayed_bband_top = self.bband1.lines.top(-1)
        self.delayed_bband_bot = self.bband1.lines.bot(-1)
        self.bband2 = bt.indicators.BollingerBands(self.dataclose, period = 20, devfactor = 3)
        ## Lower 1 - MACD
        self.macd = bt.indicators.MACDHisto(period_me1 = 6, period_me2 = 17, period_signal = 9)
        self.delayed_macd = self.macd.lines.histo(-1)
        ## Lower 2 - ADX
        self.adx = bt.indicators.AverageDirectionalMovementIndex(period = 5)
        sig = self.adx(-1)
        self.sig = self.adx.lines.adx - sig
        ## Lower 3 - Stochastic
        self.stoch = bt.indicators.Stochastic(period = 8, period_dfast = 2, period_dslow = 2)
        self.delayed_stochK = self.stoch.lines.percK(-1); self.delayed_stochK2 = self.stoch.lines.percK(-2)
        self.delayed_stochD = self.stoch.lines.percD(-1); self.delayed_stochD2 = self.stoch.lines.percD(-2)
        self.cross_stoch = bt.indicators.CrossOver(self.stoch.lines.percK, self.stoch.lines.percD)
        self.cross_stoch2 = bt.indicators.CrossOver(self.delayed_stochK, self.delayed_stochD)
        self.cross_stoch3 = bt.indicators.CrossOver(self.delayed_stochK2, self.delayed_stochD2)
    def notify_order(self, order):
        # If Buy and/or Sell submitted then accepted by broker - Nothing to do
        if order.status in [order.Submitted, order.Accepted]:
            return
        #Check if an order is completed
        if order.status in [order.Completed]:
            if self.call_put == 'buy put' and order.issell():
                self.log('BUY PUT EXECUTED, Price: {:,.2f}'.format(order.executed.price))
                self.target = order.executed.price * 0.9995
            elif self.call_put == 'sell put' and order.isbuy():
                self.log('SELL PUT EXECUTED, Price: {:,.2f}'.format(order.executed.price))
        elif order.status in [order.Cancelled, order.Margin, order.Rejected]:
            self.log('ORDER CANCELLED')
        # Write down that there is no order
        self.order = None
        return self.target
    def notify_trade(self, trade):
        if not trade.isclosed:
            return        
        self.log('OPERATION PROFIT/LOSS, GROSS: {:,.2f}'.format(trade.pnl))
    def next(self):
        # Log the next price
        self.log('Close: {:,.2f}'.format(self.dataclose[0]))
        # If there is a pending order, do not order another one
        if self.order:
            return      
        # This is for one position at a time
        # Check if there is a position in the market
        if not self.position: # Not in the market, open position
            if self.datas[0].datetime.time(0) < datetime.time(15, 30):
                def put_if():
                    # EMA
                    put_rule1 = [(self.dataclose[0] < self.ema1[0] or self.dataopen[0] < self.ema1),
                                 (self.dataclose[0] < self.ema2[0] or self.dataopen[0] < self.ema2),
                                 (self.dataclose[0] < self.ema3[0] or self.dataopen[0] < self.ema3),
                                ]
                    # MACD
                    put_rule2 = [(self.macd.lines.histo[0] < self.delayed_macd[0]),
                                  (self.macd.lines.histo[0] > -0.2),
                                  (abs(self.macd.lines.histo[0]) > 0.1),
                                 ]
                    # ADX
                    put_rule3 = [(self.adx.lines.adx[0] >= 20 and self.adx.lines.adx[0] <= 50),
                                ]
                    # Stochastic
                    put_rule4 = [(self.cross_stoch.lines.crossover == -1 or self.cross_stoch2.lines.crossover == -1\
                                  or self.cross_stoch3.lines.crossover == -1),
                                 (self.stoch.lines.percD[0] > self.stoch.lines.percK[0]),
                                 (self.stoch.lines.percD[0] >= 50 or self.delayed_stochD >= 50 or self.delayed_stochD2 >= 50),
                                 (self.stoch.lines.percK[0] >= 50 or self.delayed_stochK >= 50 or self.delayed_stochK2 >= 50),
                                 ]                
                    # Summary
                    put_rules = [put_rule1, put_rule2, put_rule3, put_rule4]
                    return put_rules
                if all(put_if()):
                    self.log('BUY PUT CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                    self.order = self.sell()
                    self.call_put = 'buy put'
        else: # Close position
            target_achieved_put = [(self.target >= self.dataclose[0]) or (self.target >= self.dataopen[0])]
            if all(target_achieved_put) and self.call_put == 'buy put':
                self.log('SELL PUT CREATE ORDER: {:,.2f}'.format(self.target))
                self.order = self.buy()
                self.call_put = 'sell put'        
```

```python
class OTi_Call(bt.Strategy):
    def log(self, txt, dt = None, h = None, doprint = True):
        '''Logging function for this strategy'''
        dt = dt or self.datas[0].datetime.date(0)
        h = h or self.datas[0].datetime.time(0)
        print('%s %s, %s' % (dt.isoformat(), h.isoformat(), txt))              
    def __init__(self): 
        # Keep a reference to close price for data[0] in the series
        self.dataclose = self.datas[0].close
        self.dataopen = self.datas[0].open
        self.datavol = self.datas[0].volume
        self.datahigh = self.datas[0].high
        self.datalow = self.datas[0].low        
        # Keep track of pending order, buy price-commission, and call-put
        self.order = None
        self.price = None
        self.buycomm = None
        self.call_put = None 
        # Indicators
        ## Upper - EMA and Bollinger
        self.ema1 = bt.indicators.EMA(self.dataclose, period = 6)
        self.ema2 = bt.indicators.EMA(self.dataclose, period = 10)
        self.ema3 = bt.indicators.EMA(self.dataclose, period = 20)
        self.bband1 = bt.indicators.BollingerBands(self.dataclose, period = 20, devfactor = 2)
        self.delayed_bband_top = self.bband1.lines.top(-1)
        self.delayed_bband_bot = self.bband1.lines.bot(-1)
        self.bband2 = bt.indicators.BollingerBands(self.dataclose, period = 20, devfactor = 3)
        ## Lower 1 - MACD
        self.macd = bt.indicators.MACDHisto(period_me1 = 6, period_me2 = 17, period_signal = 9)
        self.delayed_macd = self.macd.lines.histo(-1)
        ## Lower 2 - ADX
        self.adx = bt.indicators.AverageDirectionalMovementIndex(period = 5)
        sig = self.adx(-1)
        self.sig = self.adx.lines.adx - sig
        ## Lower 3 - Stochastic
        self.stoch = bt.indicators.Stochastic(period = 8, period_dfast = 2, period_dslow = 2)
        self.delayed_stochK = self.stoch.lines.percK(-1); self.delayed_stochK2 = self.stoch.lines.percK(-2)
        self.delayed_stochD = self.stoch.lines.percD(-1); self.delayed_stochD2 = self.stoch.lines.percD(-2)
        self.cross_stoch = bt.indicators.CrossOver(self.stoch.lines.percK, self.stoch.lines.percD)
        self.cross_stoch2 = bt.indicators.CrossOver(self.delayed_stochK, self.delayed_stochD)
        self.cross_stoch3 = bt.indicators.CrossOver(self.delayed_stochK2, self.delayed_stochD2)
    def notify_order(self, order):
        # If Buy and/or Sell submitted then accepted by broker - Nothing to do
        if order.status in [order.Submitted, order.Accepted]:
            return
        #Check if an order is completed
        if order.status in [order.Completed]:
            if self.call_put == 'buy call' and order.isbuy():
                self.log('BUY CALL EXECUTED, Price: {:,.2f}'.format(order.executed.price))
                self.target = order.executed.price * 1.001
            elif self.call_put == 'sell call' and order.issell():
                self.log('SELL CALL EXECUTED, Price: {:,.2f}'.format(order.executed.price))
        elif order.status in [order.Cancelled, order.Margin, order.Rejected]:
            self.log('ORDER CANCELLED')
        # Write down that there is no order
        self.order = None
        return self.target
    def notify_trade(self, trade):
        if not trade.isclosed:
            return        
        self.log('OPERATION PROFIT/LOSS, GROSS: {:,.2f}'.format(trade.pnl))
    def next(self):

        # Log the next price
        self.log('Close: {:,.2f}'.format(self.dataclose[0]))
        # If there is a pending order, do not order another one
        if self.order:
            return      
        # This is for one position at a time
        # Check if there is a position in the market
        if not self.position: # Not in the market, open position
            
            def call_if():
                # EMA
                call_rule1 = [(self.dataclose[0] > self.ema1[0] or self.dataopen[0] > self.ema1[0]),
                              (self.dataclose[0] > self.ema2[0] or self.dataopen[0] > self.ema2[0]),
                              (self.dataclose[0] > self.ema3[0] or self.dataopen[0] > self.ema3[0]),
                             ]
                # MACD
                call_rule2 = [(self.macd.lines.histo[0] > self.delayed_macd[0]),
                              (self.macd.lines.histo[0] > -0.2 and self.macd.lines.histo[0] < 0.2),
                              (abs(self.macd.lines.histo[0]) > 0.1),
                             ]
                # ADX
                call_rule3 = [(self.adx.lines.adx[0] >= 20 and self.adx.lines.adx[0] <= 50),
                             ]
                # Stochastic
                call_rule4 = [(self.cross_stoch.lines.crossover == 1 or self.cross_stoch2.lines.crossover == 1\
                              or self.cross_stoch3.lines.crossover == 1),
                             ]
                # Summary
                call_rules = [call_rule1, call_rule2, call_rule3, call_rule4]
                return call_rules
            if all(call_if()): 
                self.log('BUY CALL CREATE ORDER: {:,.2f}'.format(self.dataclose[0]))
                self.order = self.buy()
                self.call_put = 'buy call'
        else: # Close position
            target_achieved_call = [(self.target <= self.dataclose[0]) or (self.target <= self.dataopen[0])]
            if  all(target_achieved_call) and self.call_put == 'buy call':
                self.log('SELL CALL CREATE ORDER: {:,.2f}'.format(self.target))
                self.order = self.sell()
                self.call_put = 'sell call'
```

# Analyzers

```python
class recorder(bt.Analyzer):
    def get_analysis(self):
        return self.record, self.identity, self.trade_in
    def __init__(self):
        self.identity = WilliamR.identity
        self.trade_in = []
        self.record = []
        self.cumprofit = 0.0
    def notify_trade(self, trade):
        if trade.isopen:
            self.trade_in.append({'ref': trade.ref})
        if trade.isclosed:
            brokervalue = self.strategy.broker.getvalue() # Get portfolio value
            if True: # call-put direction
                call_put = 'put'
                if trade.history[0].event.size > 0: call_put = 'call'
            if True: # Entry-Exit price
                pricein = trade.history[len(trade.history)-1].status.price
                priceout = trade.history[len(trade.history)-1].event.price
            if True: # Entry-Exit time
                datein = bt.num2date(trade.history[0].status.dt)
                dateout = bt.num2date(trade.history[len(trade.history)-1].status.dt)
                if trade.data._timeframe >= bt.TimeFrame.Days:
                    datein = datein.date()
                    dateout = dateout.date()
                pcntchange = 100 * priceout / pricein - 100
                pnl = trade.history[len(trade.history)-1].status.pnlcomm
                pnlpcnt = 100 * pnl / brokervalue
                barlen = trade.history[len(trade.history)-1].status.barlen
                if barlen !=0:
                    pbar = pnl / barlen
                else:
                    pbar = 0
            if True: # Cumulative profit, size, and value
                self.cumprofit += pnl # Cumulative profit
                size = value = 0.0 # size and value
                for record in trade.history:
                    if abs(size) < abs(record.status.size):
                        size = record.status.size
                        value = record.status.value
            self.record.append({'date in': datein, 'date out': dateout, 'ref': trade.ref, 'call/put': call_put,
                                'ticker': trade.data._name, 'pricein': pricein, 'priceout': priceout, 
                                'size': size, 'value': value,
                                'barlength': barlen, 'change(%)': round(pcntchange, 2), 'pnl': pnl,
                                'pnl(%)': round(pnlpcnt, 2), 'pnl/bar': round(pbar, 2), 
                                'cumulative profit': self.cumprofit})
```
